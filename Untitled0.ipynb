```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data (Replace 'zomato.csv' with your actual file name)
df = pd.read_csv('zomato.csv')

# 2. See the first 5 rows to understand what's inside
print("--- First 5 Rows of Data ---")
print(df.head())

# 3. Clean the 'Rating' column (Convert '4.1/5' to 4.1)
def clean_rating(value):
    if isinstance(value, str):
        value = value.split('/')[0]
        try:
            return float(value)
        except:
            return None
    return value

# Apply the cleaning function to the rating column
# (Note: column names might be 'rate' or 'aggregate_rating' depending on your file)
df['rate'] = df['rate'].apply(clean_rating)

# 4. Create your first visualization: Number of restaurants per city
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='location', order=df['location'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Top 10 Cities with Most Restaurants on Zomato')
plt.show()
```

```python
# 1. Compare Online vs Offline Ratings
print(df.groupby('online_order')['rate'].mean())

# Convert 'approx_cost(for two people)' to numeric, handling potential errors
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

# 2. Find the "Best Value" Restaurants (High Rating + Low Cost)
# We look for Rating > 4 and Cost < 500
value_for_money = df[(df['rate'] >= 4.0) & (df['approx_cost(for two people)'] <= 500)]
print(f"Number of 'Value for Money' restaurants: {len(value_for_money)}")

# 3. Most Popular Cuisines by Vote Count
popular_cuisines = df.groupby('cuisines')['votes'].sum().sort_values(ascending=False).head(10)
print(popular_cuisines)
```

```python
# Find the city with the highest number of restaurants
top_city = df['location'].value_counts().idxmax()
print(f"Deep Diving into the top city: {top_city}")

# Create a 'subset' of data just for that city
city_df = df[df['location'] == top_city]
```

```python
# Group by locality to see where the restaurants are clustered
locality_stats = city_df.groupby('location').agg({
    'rate': 'mean',
    'votes': 'sum',
    'approx_cost(for two people)': 'mean',
    'name': 'count'
}).rename(columns={'name': 'restaurant_count'})

# Filter for localities that have at least 50 restaurants to avoid small samples
popular_localities = locality_stats[locality_stats['restaurant_count'] > 50].sort_values(by='rate', ascending=False)

print("Top 5 Localities with the highest average ratings:")
print(popular_localities.head(5))
```

```python
# Pick the top locality from your results
target_locality = popular_localities.index[0]
locality_df = city_df[city_df['location'] == target_locality]

# See the most common cuisines in this area
print(f"Cuisine distribution in {target_locality}:")
print(locality_df['cuisines'].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure visual style is clean
sns.set_theme(style="whitegrid")

# 1. Top 10 Localities (Bar Chart)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='location', order=df['location'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Localities by Restaurant Density', fontsize=15)
plt.show()

# 2. Online vs. Offline (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='online_order', y='rate', data=df, palette='Set2')
plt.title('Does Online Ordering Affect Ratings?', fontsize=15)
plt.show()

# 3. Cost vs Rating (Scatter Plot)
plt.figure(figsize=(10, 6))
# Note: 'approx_cost(for two people)' must be cleaned to numeric first!
sns.scatterplot(x='approx_cost(for two people)', y='rate', hue='online_order', data=df, alpha=0.6)
plt.title('The Sweet Spot: Cost vs. Rating', fontsize=15)
plt.show()

# 4. Top 10 Cuisines (Bar Chart)
plt.figure(figsize=(10, 6))
top_cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='magma')
plt.title('Most Popular Cuisines', fontsize=15)
plt.show()
```
